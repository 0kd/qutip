```python
from qutip.distributions import *
```

```python
N = 20
alpha = 2.0 + 2j
rho = (coherent(N, alpha) + coherent(N, -alpha)).unit()
```

### Wigner distribution

```python
w = WignerDistribution(rho, extent=[[-10, 10], [-10, 10]])
```

```python
w.visualize();
```

### Husimi Q function

```python
q = QDistribution(rho, extent=[[-10, 10], [-10, 10]])
```

```python
q.visualize();
```

### Two-mode quadrature correlations

```python
rho = tensor(coherent(N, alpha), coherent(N, -alpha))
```

```python
p = TwoModeQuadratureCorrelation(rho)
```

```python
p.visualize();
```

```python
p = TwoModeQuadratureCorrelation(ket2dm(rho))
```

```python
p.visualize();
```

```python
fig, axes = subplots(1, 2, figsize=(12, 5))

w0 = WignerDistribution(ptrace(rho, 0))
w1 = WignerDistribution(ptrace(rho, 1))

w0.visualize(fig=fig, ax=axes[0]);
w1.visualize(fig=fig, ax=axes[1]);
```

### Marginal distribution

```python
alpha = 2.0
rho = (coherent(N, alpha) + coherent(N, -alpha)).unit()
```

```python
w = WignerDistribution(rho)
```

```python
w.visualize();
```

```python
wx = w.marginal(dim=0)
```

```python
fig, ax = wx.visualize();
```

```python
wy = w.marginal(dim=1)
```

```python
fig, ax = wy.visualize();
```

```python
wx = w.project(dim=1)
```

```python
fig, ax = wx.visualize();
```

## Visualization

```python
def plot_correlation(taulist, e_list, corr_list):
    
    fig, axes = subplots(2, 1, sharex=True, figsize=(10,4))

    for idx, corr in enumerate(corr_list):
        axes[0].plot(taulist, real(corr), label="correlation %d" % idx)

    for idx, e in enumerate(e_list):
        axes[1].plot(taulist, real(e), label="expectation %d" % idx)

    axes[0].legend()
    axes[1].legend()
    axes[1].set_xlabel(r'$\tau$');
```

## Model problem: Harmonic oscillator

```python
N = 20
w0 = 2 * pi
G1 = 0.75
n_th = 2.00
taulist = linspace(0, 10.0, 500)

# operators and hamiltonian 
a = destroy(N)
H = w0 * a.dag() * a
c_ops = [sqrt(G1*(1+n_th)) * a, sqrt(G1*n_th) * a.dag()]

# initial state
rho0 = thermal_dm(N, 4.0)
rho0 = coherent_dm(N, sqrt(4.0))
```

## operator $a$

```python
A = a.dag()
B = a
n = mesolve(H, rho0, taulist, c_ops, [A * B]).expect[0]
corr = correlation(H, rho0, None, taulist, c_ops, A, B)
plot_correlation(taulist, [n], [corr])
```

```python
corr = correlation(H, rho0, None, taulist, c_ops, A, B, reverse=True)
plot_correlation(taulist, [n], [corr])
```

### operator $x$

```python
A = a + a.dag()
B = A
n = mesolve(H, rho0, taulist, c_ops, [A * B]).expect[0]
corr = correlation_ss(H, taulist, c_ops, A, B, rho0=rho0)
plot_correlation(taulist, [n], [corr])
```

```python
corr = correlation_ss(H, taulist, c_ops, A, B, rho0=rho0, reverse=True)
plot_correlation(taulist, [n], [corr])
```

### operator $p$

```python
A = -1j*(a + a.dag())
B = A.dag()
n = mesolve(H, rho0, taulist, c_ops, [A * B]).expect[0]
corr = correlation_ss(H, taulist, c_ops, A, B, rho0=rho0)
plot_correlation(taulist, [n], [corr])
```

```python
corr = correlation_ss(H, taulist, c_ops, A, B, rho0=rho0, reverse=True)
plot_correlation(taulist, [n], [corr])
```

### operator $i a$

```python
A = 1j * a
B = a.dag()
n = mesolve(H, rho0, taulist, c_ops, [A * B]).expect[0]
corr = correlation_ss(H, taulist, c_ops, A, B, rho0=rho0)
plot_correlation(taulist, [n], [corr])
```

```python
corr = correlation_ss(H, taulist, c_ops, A, B, rho0=rho0, reverse=True)
plot_correlation(taulist, [n], [corr])
```

## Test and compare different solvers

```python
A = a.dag()
B = a

A = 1j * a
B = a.dag()

n = mesolve(H, rho0, taulist, c_ops, [A * B]).expect[0]

corr1 = correlation(H, rho0, None, taulist, c_ops, A, B, solver="me", reverse=True)
corr2 = correlation(H, rho0, None, taulist, c_ops, A, B, solver="es", reverse=True)

plot_correlation(taulist, [corr1], [corr2])
```
